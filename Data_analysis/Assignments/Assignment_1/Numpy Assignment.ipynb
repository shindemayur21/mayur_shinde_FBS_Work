```python
# 2. Create array with 20 random integers from 1 to 50
arr = np.random.randint(1, 50, 20)

print("Original Array:")
print(arr)

# Replace odd numbers with -1
arr[arr % 2 != 0] = -1

print("\nModified Array:")
print(arr)
```

```python
# 3. Perform the following operations on an array of mobile phones prices 6999, 7500,
# 11999, 27899, 14999, 9999.
# a. Create a 1d-array of mobile phones prices

prices = np.array([6999, 7500, 11999, 27899, 14999, 9999])

print("Original Array:")
prices
```

```python
# b. Convert this array to float type
prices = prices.astype(float)

print("\nFloat Type Array:")
prices
```

```python
# c. Append a new mobile price
prices = np.append(prices, 13999)

print("\nAfter Appending 13999:")
prices
```

```python
# d. Reverse the array
prices = prices[::-1]

print("\nReversed Array:")
prices
```

```python
# f. Sort in descending order
prices = np.sort(prices)[::-1]

print("\nSorted in Descending Order:")
prices
```

```python
# g. Average mobile phone price
average_price = np.mean(prices)

print("\nAverage Price:")
average_price
```
