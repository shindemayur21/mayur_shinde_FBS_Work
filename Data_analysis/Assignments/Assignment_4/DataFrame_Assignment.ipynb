```python
import pandas as pd
```

```python
# 1. Create a DataFrame:
# Load the above dataset into a Pandas DataFrame.

sl = pd.DataFrame({'Order ID' : [1001, 1002, 1003, 1004, 1005, 1006, 1007, 1008],
                   'Customer' : ['Alice', 'Bob', 'Charlie', 'Alice', 'David', 'Charlie', 'Bob', 'Alice'],
                   'Product' : ['Laptop', 'Phone', 'Chair', 'Table', 'Laptop', 'Phone', 'Chair', 'Phone'],
                   'Category' : ['Electronics', 'Electronics', 'Furniture', 'Furniture', 'Electronics', 'Electronics', 'Furniture', 'Electronics'],
                   'Quantity' : [2, 1, 4, 1, 1, 2, 2, 1],
                   'Price' : [800, 500, 100, 250, 800, 500, 100, 500],
                   'Total Sales' : [1600, 500, 400, 250, 800, 1000, 200, 500]})
sl
```

```python
# 2. Add a New Column:
# Add a column Discount that gives a 10% discount on total sales.

sl['Discount'] = sl['Total Sales'] * 0.10
sl
```

```python
# 3. Add a New Row:
# Add a new sales record: Order ID = 1009, Customer =
# "Eve", Product = "Tablet", Category = "Electronics",
# Quantity = 1, Price = 300, Total Sales = 300.

sl.loc[len(sl)] = [1009, 'Eve', 'Tablet', 'Electronics', 1, 300, 300, 300 * 0.10]
sl
```

```python
# 4. Filter Data:
# Retrieve only the sales records where the Category is "Electronics".

s2 = sl.loc[sl['Category'] == 'Electronics']
s2
```

```python
# 5. Selection:
# Select the Customer and Total Sales columns.

s3 = sl[['Customer', 'Total Sales']]
s3
```

```python
# 6. Grouping:
# Group the data by Customer and calculate the total sales per customer.

grps = sl.groupby('Customer')['Total Sales'].sum()
grps
```

```python
# 7. Sorting:
# Sort the DataFrame based on Total Sales in descending order.

srt = sl.sort_values(by='Total Sales')
srt
```

```python
# 8. Apply Function:
# Apply a lambda function to create a new column Sales Category,
# where values are labeled as "High" if Total Sales > 1000, otherwise "Low".
    
sl['Sales Category'] = sl['Total Sales'].apply(lambda x: 'High' if x > 750 else 'low')
sl
```

```python
# 9. Save DataFrame:
# Save the final DataFrame as a CSV file named sales_data.csv.
sl.to_csv('sales_data.csv', index=False)
```
